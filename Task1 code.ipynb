```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('http://bit.ly/w-data')
print(df)
```

```python
#we will plot a graph between the two to see the kind of relation between them
sns.scatterplot(data=df, x="Scores", y="Hours")
#the above graph suggest a linear relation
```

```python
sns.distplot(df['Scores'], label = 'Skewness: %.2f'%(df['Hours'].skew()))
#the above graph shows there is no skew in the data
```

```python
from sklearn.model_selection import train_test_split

hoursX = df['Hours'] # or df.iloc[:, :-1].values
scoreY = df['Scores'] #or df.iloc[:, 1].values
X_train, X_test, y_train, y_test = train_test_split(hoursX, scoreY, test_size=0.185, random_state=2 )
print('s')
```

```python
from sklearn.linear_model import LinearRegression

X_train = X_train.values.reshape(-1,1) #reshape not needed if we use the iloc[:, :-1].values method
X_test = X_test.values.reshape(-1,1)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
```

```python
sns.lmplot(x = 'Hours', y = 'Scores', data = df)
```

```python
#compare
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
time = 9.25
aim = regressor.predict([[time]])
print('If a student studies for {} hours/day.'.format(time))
print('He will get {} marks.'.format(aim[0]))
```
